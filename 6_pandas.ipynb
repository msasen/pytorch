# Pandas Series
Bir db'deki tablonun entity'leri gibi herhangi bir tür data'dan birden çok veri tutan yapılardır

```python
import pandas as pd

a = [1, 7, 2]

pd.Series(a)
```

Başka bir şey belirtilmezse, değerler dizin numaralarıyla etiketlenir. İlk değerin indeksi 0, ikinci değerin indeksi 1 vb.

Bu etiket, belirtilen bir değere erişmek için kullanılabilir.



```python
myvar[0]
```

Kendi özel etiketinizi oluşturabilirsiniz.

```python
myvar = pd.Series(a, index = ["x", "y", "z"])
myvar
```

Oluşturduğunuz etiketi bu şekilde kullanabilirsiniz.

```python
myvar["y"]
```

Dic türü pandas series'e aşağıdaki gibi dönüştürülebilir.

```python
calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories)
myvar
```

Sadece day1 ve day2'yi dahil etmek istediğimizde oluşan durum aşağıdaki gibidir.

```python
calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories, index = ["day1", "day2"])

print(myvar)
```
